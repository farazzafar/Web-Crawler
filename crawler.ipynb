```python
import requests
from bs4 import BeautifulSoup
```

```python
def getpage(url):
    try:
        page = requests.get(url)
        if page.status_code == 200:
            print('success')
            return page.text
        else:
            print('error loading page')
            
    except Exception as e:
        print(e)
            
```

```python
data = getpage('https://www.google.com/search?q=times+of+india')
```

```python
soup = BeautifulSoup(data, 'lxml')
```

```python
soup.find('h1').text
```

```python
soup.find_all('h3')
```

```python
for item in soup.find_all('ul'):
    print(item.text)
```

```python
paras = soup.find_all('ul',attrs={'data-vr-zone':'top_stories'})
```

```python
for item in paras:
    print(item.text)
```
